### Integral arrays (prefix sums)

$a = [a_1, a_2, a_3, ..., a_n]$

$\sum_{i=k}^m a_i$

$s = [0, a_1, a_1 + a_2, a_1 + a_2 + a_3, ..., a_1 + ... + a_n]$

$\sum_{i=k}^m a_i = s[m+1] - s[k]$

```python
import time

def calc_integral_arr(a):
    s = 0
    int_arr = [s]
    
    for el in a:
        s += el
        int_arr.append(s)
    
    return int_arr

def subarray_sum_naive(a, k, m):
    s = 0
    
    for el in a[k : m + 1]:
        s += el
    
    return s

def subarray_sum_integral(iarr, k, m):
    return iarr[m + 1] - iarr[k]

a = [1, 2, 3, 4]

integral = calc_integral_arr(a)

#print(integral)

print(subarray_sum_naive(a, 0, len(a) - 1))

print(subarray_sum_integral(integral, 0, len(a) - 1))
```

```python
inp_len = 10000
op_num = 10000

inp = [i for i in range(inp_len)]

start_ind, end_ind = inp_len // 4, inp_len * 3 // 4

start_time = time.time()

for _ in range(op_num):
    subarray_sum_naive(inp, start_ind, end_ind)

finish_time = time.time()

print(f"{op_num} naive operations time elapsed: {finish_time - start_time}")

start_time = time.time()

integral_array = calc_integral_arr(inp)

for _ in range(op_num):
    subarray_sum_integral(integral_array, start_ind, end_ind)

finish_time = time.time()

print(f"{op_num} operations with prefix sums time elapsed: {finish_time - start_time}")
```

```python
#Продолжите её, добавив рекурсивную функцию изменения k-го элемента списка.
# Это можно сделать в рамках того же класса Node.
# Создайте функцию, которая будет вызывать себя же, но у своего next, и менять значение элемента, если до него дошла.
# Также реализуйте функцию рекурсивной распечатки односязного списка.

class Node:
    def __init__(self, val, name = "noname"):
        self.next = None
        self.val  = val
        self.name = name
        #print(f"val = {self.val}")
    
    def get_val(self):
        return self.val

    def get_next(self):
        return self.next

    def get_name(self):
        return self.name
    
    def set_next(self, new_next):
        self.next = new_next
    
    def print_recursive(self):
        string = str(self.val)
        
        if (self.next is not None):
            string = string + " -> " + self.next.print_recursive()
        
        return string
    
    def change_val(self, k, new_val):
        if (k == 0):
            self.val = new_val
        
        elif (self.next is not None):
            self.next.change_val(k - 1, new_val)
        
        else:
            print(f"required index out of range! cannot change value")
    
    #def print_iterative(self):
        

# class Single_linked_list:
#     def __init__(self):
#         self.root = None

#single-linked list

n1 = Node(6, "first")
n2 = Node(7, "second")
n3 = Node(9, "third")
n4 = Node(12, "afirst")
n5 = Node(17, "asecond")
n6 = Node(19, "athird")

#print(f"val n2 = {n2.get_val()}")

n1.set_next(n2)
n2.set_next(n3)
n3.set_next(n4)
n4.set_next(n5)
n5.set_next(n6)

#print(f"val n1 = {n1.get_val()}")
#print(f"val n2 = {n1.next.get_val()}")
#print(f"val n3 = {n1.next.next.get_val()}")

#print(f"first name: {n1.next.next.next.next.get_name()}")

print(n1.print_recursive())

n1.change_val(15, 15)

print(n1.print_recursive())
```

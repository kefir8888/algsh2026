```python
class Node:
    def __init__(self, val, name = "noname"):
        self.next = None
        self.val  = val
        self.name = name
        print(f"val = {self.val}")
    
    def get_val(self):
        return self.val

    def get_next(self):
        return self.next

    def get_name(self):
        return self.name
    
    def set_next(self, new_next):
        self.next = new_next

class Single_linked_list:
    def __init__(self):
        self.root = None

#single-linked list

n1 = Node(6, "first")
n2 = Node(7, "second")
n3 = Node(9, "third")
n4 = Node(12, "afirst")
n5 = Node(17, "asecond")
n6 = Node(19, "athird")

#print(f"val n2 = {n2.get_val()}")

n1.set_next(n2)
n2.set_next(n3)
n3.set_next(n4)
n4.set_next(n5)
n5.set_next(n6)

#print(f"val n1 = {n1.get_val()}")
#print(f"val n2 = {n1.next.get_val()}")
#print(f"val n3 = {n1.next.next.get_val()}")

print(f"first name: {n1.next.next.next.next.get_name()}")
```

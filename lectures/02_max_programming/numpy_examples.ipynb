### Немного основных функций numpy

```python
import numpy as np

#cоздание массивов

#массов из единиц 3x5, умноженный на 9. Тип данных - беззнаковые (unsigned) целые (integer) числа
arr = np.ones((3, 5), np.uint8)

print(arr)
```

```python
#shape - "форма", то есть фактически размеры
arr.shape
```

```python
#можно записать в массив число в отдельный элемент
arr[2, 4] = 52
print(arr)
```

```python
#или в целый подмассив
arr[1:3, 2:4] = 96
print(arr)
```

```python
#ну там сумма, среднее
print(np.sum(arr), np.mean(arr))
```

```python
#если не только про numpy говорить, то можно еще рисовать массивы

import matplotlib.pyplot as plt

plt.imshow(arr)
```

```python
#тут по умолчанию такая цветовая схема. Можно поставить оттенки серого. Все пятьесят (зачеркнуто) двести пятьдесят шесть штук

plt.imshow(arr, cmap="gray")
```

```python
#%%time

import numpy as np
import time
import matplotlib.pyplot as plt

def find_max(arr):
    for _ in range(3000):
        if (len(arr) == 0):
            return "hehehe"
        
        m = arr[0]
        
        for el in arr[1:]:
            m = max(m, el)    
    
    if (len(arr) == 0):
        return "hehehe"
    
    m = arr[0]
    
    for el in arr[1:]:
        m = max(m, el)
    
    return m

def find_slow_max(arr):
    n = len(arr)
    
    comparison_table = np.zeros((n, n))
    
    #comparison_table[5, 2] = 1
        
    #fill the table
    
    for i in range(n):
        for j in range(n):
            if (arr[i] >= arr[j]):
                comparison_table[i, j] = 1
    
    #print(comparison_table)

    #find maximum
    
    sums = np.sum(comparison_table, axis=1, keepdims = True)
    
    #print(sums)
        
    for i in range(n):
        if (sums[i, 0] == n):
            return arr[i]

def test_max(max_func, step, stepnum):
    nums  = []
    times = []
    
    for i in range(stepnum):
        input_size = step * (i + 1)
        nums.append(input_size)
        
        arr = [np.random.randint(1001) for _ in range(input_size)]

        start_time = time.time()
        m = max_func(arr)
        finish_time = time.time()
        
        #print(f"max el: {m}")
        time_elapsed = finish_time - start_time
        times.append(time_elapsed)
    
    return nums, times

funcs = [max, find_max, find_slow_max]

for func in funcs:
    nums, times = test_max(func, 100, 20)

    #print(nums, times)

    plt.plot(nums, times)
```

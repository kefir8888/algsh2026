```python
def merge(l, r):
    merged = []
    
    i, j = 0, 0
    
    while(True):
        if (i == len(l) and j == len(r)):
            break
        
        elif (i == len(l)):
            merged.append(r[j])
            j += 1
        
        elif (j == len(r)):
            merged.append(l[i])
            i += 1
        
        else:
            if (l[i] < r[j]):
                merged.append(l[i])
                i += 1
            
            else:
                merged.append(r[j])
                j += 1
    
    return merged

# arr1 = [2, 2, 3, 5, 5, 7, 8, 9]
# arr2 = [3, 4, 5, 5, 6, 7]

# merged = merge(arr1, arr2)

# print(merged)

def merge_sort(arr, depth = 0):
    print(f"inp: {depth} : {arr}")
    
    if (len(arr) < 2):
        return arr
        
    m = len(arr) // 2
    
    l = merge_sort(arr[:m], depth + 1)
    r = merge_sort(arr[m:], depth + 1)
    
    merged = merge(l, r)
    
    print(f"out : {depth} : {merged}")
    
    return merged

arr = [5, 3, 7, 6, 7, 9, 5, 4, 2, 6, 4, 3, 2]

print(merge_sort(arr))
```

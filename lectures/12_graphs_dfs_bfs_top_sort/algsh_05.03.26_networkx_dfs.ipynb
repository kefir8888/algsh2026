#### Depth-First Search

```python
import matplotlib.pyplot as plt
import networkx as nx

g = nx.Graph()

g.add_node(0)
g.add_node(1)
g.add_node(2)
g.add_node(3)
g.add_node(4)
g.add_node(5)

g.add_edge(0, 3)
g.add_edge(1, 3)
g.add_edge(1, 2)
g.add_edge(1, 4)
g.add_edge(2, 3)
g.add_edge(2, 4)
g.add_edge(4, 5)
g.add_edge(2, 5)

pos = nx.spring_layout(g)

def dfs(g, v, color_map, pos, num):
    plt.figure(num)
    nx.draw(g, node_color=color_map, with_labels=True, pos = pos)
    num += 1
    
    #check if current node is already closed
    if (color_map[v] != "red"):
        return
    
    #change color
    color_map[v] = "orange"
    plt.figure(num)
    nx.draw(g, node_color=color_map, with_labels=True, pos = pos)
    num += 1
    
    #recursive call
    color_map[v] = "yellow"
    agjacent = list(g[v].keys())
    
    for node in agjacent:
        print(f"current {v}, calling at {node}")
        dfs(g, node, color_map, pos, num)
        
    #close node
    color_map[v] = "green"
    
    plt.figure(num)
    nx.draw(g, node_color=color_map, with_labels=True, pos = pos)
    num += 1

def dfs_full_graph(g, pos):
    color_map = ["red" for i in range(len(g.nodes()))]
    
    dfs(g, 1, color_map, pos, 0)
    
    print("dfs complete")
    
    nx.draw(g, node_color=color_map, with_labels=True, pos = pos)

dfs_full_graph(g, pos)
```

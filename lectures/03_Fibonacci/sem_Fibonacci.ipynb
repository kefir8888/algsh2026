```python
import time

def Fk_def(k):
    if (k < 2):
        return k
    
    return Fk_def(k-1) + Fk_def(k-2)

last_time_elapsed = 1
last_Fk = Fk_def(1)

times = []

for k in range(2, 37, 1):
    start_time = time.time()
    Fk = Fk_def(k)
    finish_time = time.time()
    time_elapsed = finish_time - start_time
    
    print(k, Fk, time_elapsed, time_elapsed / last_time_elapsed, Fk / last_Fk)
    last_time_elapsed = time_elapsed
    last_Fk = Fk
    
    times.append(time_elapsed)
```

```python
import matplotlib.pyplot as plt

plt.plot(times)
```

```python
import time

def Fk_def(k):
    if (k < 2):
        return k
    
    return Fk_def(k-1) + Fk_def(k-2)

def Fk_mem(k):
    Fs = [0, 1]
    
    for _ in range(2, k - 1):
        Fs.append(Fs[-1] + Fs[-2])
    
    return Fs[-1]

def test_Fibonacci(fib_func, K, k_step = 1):
    times = []

    for k in range(0, K, k_step):
        start_time = time.time()
        Fk = fib_func(k)
        finish_time = time.time()
        time_elapsed = finish_time - start_time
        
        times.append(time_elapsed)
    
    return times

K = 35

funcs = [Fk_def, Fk_mem]

for func in funcs:
    times = test_Fibonacci(func, K)
    plt.plot(times)
```

```python
funcs = [Fk_mem]

for func in funcs:
    times = test_Fibonacci(func, K = 100000, k_step = 1000)
    plt.plot(times)
```
